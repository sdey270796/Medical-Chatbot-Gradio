# Problem Statement

Fibromyalgia is a long-term, chronic condition characterized by widespread pain, extreme fatigue, and a variety of cognitive and physical symptoms. Due to its complex nature and lack of a clear diagnostic test, many patients struggle to receive proper diagnosis and treatment. The problem is further compounded by the need for effective management of symptoms to improve quality of life, as there is currently no cure for fibromyalgia.

# Objectives

1. **Raise Awareness and Understanding**: To educate patients and caregivers about fibromyalgia, its symptoms, and its impact on daily life through detailed informational resources.
2. **Improve Diagnosis and Treatment**: To provide healthcare professionals with the knowledge necessary to diagnose fibromyalgia through symptom evaluation and rule out other conditions.
3. **Explore Treatment Options**: To review the various available treatments, including physical therapies, psychological therapies, and drug treatments, and explain how each can help manage symptoms.
4. **Encourage Self-Management**: To empower individuals with fibromyalgia to take control of their condition through self-care techniques such as exercise, diet, and relaxation strategies.
5. **Promote Research and New Developments**: To support ongoing research into the causes, effects, and potential treatments of fibromyalgia in order to find more effective solutions.

# Impacts

1. **Improved Patient Well-Being**: Educating individuals on fibromyalgia can lead to better symptom management, improving the overall quality of life.
2. **Reduction in Misdiagnosis**: Providing healthcare professionals with comprehensive information on fibromyalgia's unique symptoms can reduce the chances of misdiagnosis and ensure patients receive appropriate treatment earlier.
3. **Better Symptom Control**: By promoting physical activity, psychological support, and self-care, patients may experience reduced pain, fatigue, and cognitive issues, improving their ability to perform daily activities.
4. **Greater Support for Research**: Encouraging ongoing research and new treatments will contribute to the scientific understanding of fibromyalgia, ultimately leading to better care for future patients.

```python
embeddings = HuggingFaceEmbeddings(model_name = embedding_model_path,
                                   model_kwargs = {'device':'cuda'},
                                   encode_kwargs = {'normalize_embeddings':True})
```

#Loading the PDF

```python
from langchain.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
pdf_reader = PyPDFLoader('/content/Models/GenAI Assignment/fibromyalgia-information-booklet-july2021.pdf')
```

```python
document = pdf_reader.load()
```

## Creating Chunks from PDF

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size = 1000,
                                               chunk_overlap = 200,
                                               length_function = len)
```

```python
chunk = text_splitter.split_documents(document)
```

## Passing chunks and embeddings to vectorstore

```python
from langchain.vectorstores import FAISS
```

```python
vector_db = FAISS.from_documents(documents = chunk,
                                 embedding = embeddings)
```

# Creating a retriever

```python
retriever = vector_db.as_retriever()
```

#Initializing a template

```python
template = """Given the following conversation and a follow up question, rephrase the follow up question to be a standalone question.
Chat History:
{chat_history}
Follow up Input: {question}
Standalone questions:"""
```

```python
from langchain.chains import ConversationalRetrievalChain
from langchain.prompts import PromptTemplate
```

```python
CONDENSE_QUESTION_PROMPT = PromptTemplate.from_template(template)
```

```python
chain = ConversationalRetrievalChain.from_llm(llm = llm,
                                              retriever = retriever,
                                              condense_question_prompt = CONDENSE_QUESTION_PROMPT,
                                              return_source_documents = True,
                                              verbose=False)
```

#Asking Query from User

```python
chat_history = []
query = "What are the main symptoms of fibromyalgia?"
result = chain.invoke({"question": query, "chat_history": chat_history})
print(result["answer"])
```

```python
from langchain.schema import AIMessage, HumanMessage
```

```python
chat_history.extend([HumanMessage(content=query),
                     AIMessage(content = result["answer"])])
```

## Load the Models

```python
import gradio as gr
from langchain.llms import CTransformers
from sentence_transformers import SentenceTransformer

llm = CTransformers(model="/content/Models/GenAI Assignment/llama-2-7b-chat.ggmlv3.q4_0.bin",
                    model_type="llama",
                    config={'max_new_tokens': 600,
                            'temperature': 0.01,
                            'context_length': 5000})

retriever_model = SentenceTransformer('all-MiniLM-L6-v2')
```

## Define the Inference Function

```python
def rag_inference(query):
    # Embedding the query using SentenceTransformer
    query_embedding = retriever_model.encode(query)

    # Using the LLaMA model for text generation based on the query
    response = llm(query)

    return response
```

### Gradio Interface

```python
interface = gr.Interface(fn=rag_inference,
                         inputs="text",  # Input type is text
                         outputs="text", # Output is also text
                         title="Medical Chatbot",
                         description="Ask a question, and the model will generate an answer using RAG.")
```

## Launch the Gradio App

```python
interface.launch()
```
